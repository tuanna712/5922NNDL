```python
# Perceptron Learning Algorithm

import numpy as np
import matplotlib.pyplot as plt

def generate_data(n):
    # Generate n random points in the 2D plane
    X = np.random.uniform(-1, 1, (n, 2))
    # Generate the line that separates the points
    x1, y1, x2, y2 = np.random.uniform(-1, 1, 4)
    a = (y2 - y1) / (x2 - x1)
    b = y1 - a * x1
    # Compute the labels
    y = np.sign(X[:, 1] - a * X[:, 0] - b)
    return X, y, a, b

def pla(X, y):
    # Initialize the weights
    w = np.zeros(3)
    # Add the bias term
    X = np.c_[np.ones(X.shape[0]), X]
    # Initialize the number of iterations
    it = 0
    # Iterate until all points are correctly classified
    while np.any(y != np.sign(np.dot(X, w))):
        # Find the first misclassified point
        i = np.where(y != np.sign(np.dot(X, w)))[0][0]
        # Update the weights
        w += y[i] * X[i]
        # Increment the number of iterations
        it += 1
    return w, it

def plot_data(X, y, a, b, w):
    # Plot the points
    plt.scatter(X[y == 1, 0], X[y == 1, 1], c='blue', marker='o')
    plt.scatter(X[y == -1, 0], X[y == -1, 1], c='red', marker='x')
    # Plot the line that separates the points
    x = np.linspace(-1, 1, 100)
    plt.plot(x, a * x + b, label='Target function')
    plt.plot(x, -w[1] / w[2] * x - w[0] / w[2], label='Learned function')
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.legend()
    plt.show()

def main():
    # Generate the data
    X, y, a, b = generate_data(100)
    print('Target function:', 'y =', a, '* x +', b)
    # Apply the Perceptron Learning Algorithm
    w, it = pla(X, y)
    # Plot the data
    plot_data(X, y, a, b, w)
    print('Number of iterations:', it)
```
